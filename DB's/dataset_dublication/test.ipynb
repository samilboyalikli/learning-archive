````python
# Load your original CSV file
df = dd.read_csv('dataset/credits.csv')
df.head()
````

````python
# Duplicate the DataFrame 100 times (delayed operation)
large_df = df.map_partitions(lambda pdf: pd.concat([pdf] * 300))

# Save the enlarged DataFrame to a new CSV file
large_df.to_csv('large_dataset.csv', index=False, single_file=True)

# Trigger the computation
large_df.compute()
````
